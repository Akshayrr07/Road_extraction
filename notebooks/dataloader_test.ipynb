```python
import sys
from pathlib import Path
sys.path.append("../training")
from dataset import RoadDataset
```

```python
IMAGE_DIR = "../dataset/raw/train/images"
MASK_DIR = "../dataset/raw/train/masks"
```

```python
dataset = RoadDataset(
    image_dir=IMAGE_DIR,
    mask_dir=MASK_DIR,
    image_size=256
)
print("Dataset Size:", len(dataset))
```

```python
image, mask = dataset[0]
```

```python
print("\nIMAGE")
print("Shape:", image.shape)
print("Datatype:", image.dtype)
print("Min:", image.min().item())
print("Max:", image.max().item())
```

```python
print("\nMASK")
print("Shape:", mask.shape)
print("Datatype:", mask.dtype)
print("Unique Values:", mask.unique())
```

```python
import matplotlib.pyplot as plt

# Convert tensor image back for display
image_display = image.permute(1, 2, 0).numpy()
mask_display = mask.squeeze().numpy()

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(image_display)
plt.title("Processed Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(mask_display, cmap="gray")
plt.title("Processed Mask")
plt.axis("off")

plt.show()
```

```python
from torch.utils.data import DataLoader

loader = DataLoader(
    dataset,
    batch_size=4,
    shuffle=True
)

images, masks = next(iter(loader))

print("Batch Image Shape:", images.shape)
print("Batch Mask Shape:", masks.shape)
```
