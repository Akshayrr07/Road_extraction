```python
from pathlib import Path
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
IMAGE_DIR = Path("../dataset/raw/train/images")
MASK_DIR = Path("../dataset/raw/train/masks")
```

```python
image_paths = sorted(list(IMAGE_DIR.glob("*")))
mask_paths = sorted(list(MASK_DIR.glob("*")))
```

```python
print("Total Images:", len(image_paths))
print("Total Masks :", len(mask_paths))
```

```python
sample_image = cv2.imread(str(image_paths[0]))
sample_mask = cv2.imread(str(mask_paths[0]), cv2.IMREAD_GRAYSCALE)
```

```python
print("\nIMAGE INFO")
print("Shape :", sample_image.shape)
print("Datatype :", sample_image.dtype)
```

```python
print("\nMASK INFO")
print("Shape :", sample_mask.shape)
print("Datatype :", sample_mask.dtype)
```

```python
print("\nMASK UNIQUE VALUES")
print(np.unique(sample_mask))
```

```python
sample_image_rgb = cv2.cvtColor(sample_image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(sample_image_rgb)
plt.title("Satellite Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(sample_mask, cmap="gray")
plt.title("Road Mask")
plt.axis("off")

plt.show()
```
